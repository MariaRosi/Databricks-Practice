# Table Create
## Using Pyspark

```python
spotify_df = spark.read.csv("/Volumes/datacentre/bronze/input_files/spotify_songs.csv", header=True, inferSchema=True)

display(spotify_df)
```

```python
spotify_df.write.mode("overwrite").saveAsTable("datacentre.bronze.spotify_songs_pyspark")
```

```python
%sql
   select * from datacentre.bronze.spotify_songs_pyspark
```

## Using SQL

```python
%sql
CREATE TABLE IF NOT EXISTS datacentre.bronze.spotify_songs_sql
AS
SELECT
  *,
  current_timestamp() AS ingestion_date
FROM
  read_files(
    '/Volumes/datacentre/bronze/input_files/spotify_songs.csv',
    FORMAT => 'CSV'
  )
```

```python
%sql
select * from datacentre.bronze.spotify_songs_sql
```
