```python

data = [("James","Sales","NY",90000), ("Michael","Sales","NY",76000),("Robert","Sales","CA",86000), ("Maria","Finance","CA",99000)]

schema = ["emp_name","dept","state","salary"]


    
```

```python
emp_df = spark.createDataFrame(data,schema)

display(emp_df)
```

```python
emp_df.write.format("delta").mode("overwrite").saveAsTable("employee_details.default.employee")
```

```python
from delta.tables import *
```

```python
DeltaTable.create(spark) \
    .tableName("employee_details.default.employee_new") \
    .addColumn("emp_name", "STRING") \
    .addColumn("dept", "STRING") \
    .addColumn("state", "STRING") \
    .addColumn("salary", "LONG") \
    .execute()
```
