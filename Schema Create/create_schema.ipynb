## Create Table

```python
%sql
create or replace table datacentre.bronze.customer_reviews (
    customer_id int,
    first_name string,
    last_name string,
    review_comment string
);
```

```python
%sql
insert into datacentre.bronze.customer_reviews values
(1, 'John', 'Doe', 'Amazing product! I love it.'),
(2, 'Jane', 'Smith', 'Not good, I am disappointed.'),
(3, 'Alice', 'Brown', 'Decent quality but can be better.'),
(4, 'Bob', 'Johnson', 'Fantastic service! Highly recommend.'),
(5, 'Charlie', 'Davis', 'Terrible experience, never buying again.');
```

```python
%sql
select * from datacentre.bronze.customer_reviews;
```

## Functions

```python
%sql
create function datacentre.bronze.full_name_udf(firstname string, lastname string)
returns string
return firstname || ' ' || lastname;
```

```python
%sql
select 
  *,
  datacentre.bronze.full_name_udf(first_name, last_name) as full_name
from datacentre.bronze.customer_reviews
```

```python
%sql
CREATE FUNCTION datacentre.bronze.sentiment_analysis(review_comment STRING)
RETURNS STRING
RETURN
  CASE
    WHEN review_comment LIKE '%amazing%' OR review_comment LIKE '%fantastic%' OR review_comment LIKE '%love%' OR review_comment LIKE '%recommend%' THEN 'Positive'
    WHEN review_comment LIKE '%not good%' OR review_comment LIKE '%disappointed%' OR review_comment LIKE '%terrible%' OR review_comment LIKE '%never buying%' THEN 'Negative'
    ELSE 'Neutral'
  END
```

```python
%sql
select
  *,
  datacentre.bronze.full_name_udf(first_name,last_name) as full_name,
  datacentre.bronze.sentiment_analysis(review_comment) as customer_sentiment
from datacentre.bronze.customer_reviews
```
