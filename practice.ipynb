```python
emp_df = spark.read.format("csv").option("inferSchema", "true").option("header", "true").option("sep",",").load("/Volumes/datacentre/bronze/input_files/employee_large.csv")

display(emp_df)
```

## Column Select

```python
display(emp_df.select('department').distinct())
```

```python
emp_df.select("name","department").display()
```

```python
from pyspark.sql.functions import col
emp_df.select(col("name"), col("department"))
```

## Filter

```python
emp_df_filtered = emp_df.filter(((emp_df.department == 'IT') | (emp_df.department == 'Operations')) & (emp_df.salary > 50000))
```

```python
emp_df.filter("department = 'IT' OR department = 'Operations' AND salary > 50000").display()
```

```python
display(emp_df_filtered)
```

```python
emp_df.filter(emp_df["department"].isin("IT","HR")).display()
```

```python
emp_df_filtered.groupBy('department').count().display()
```

```python
emp_df_filtered.groupBy('department').sum('salary').display()
```
